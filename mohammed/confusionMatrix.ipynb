```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

variable_names1 = ['as Track', 'as Shower']
variable_names2 = ['True Track', 'True Shower']


confusion = np.array([[0.96394287, 0.03605713],
                     [0.07943827, 0.92056173]])
annot_arr = np.array([['0.964±0.002','0.04±0.01'],
                      ['0.08±0.01','0.921±0.004']], dtype = str)

plt.figure(figsize = (5,1.3))
heatmap = sns.heatmap(confusion, annot=annot_arr, cmap='coolwarm', fmt="s", linewidths=.5, square=False, annot_kws={"size": 12})
heatmap.set_xticklabels(variable_names1, rotation=0, fontsize=12)
heatmap.set_yticklabels(variable_names2, rotation=0, fontsize=12)
#plt.title("Confusion Matrix")±±±±±
plt.tight_layout()
plt.savefig('confusionMatrixPerfect.png', dpi=500)
plt.show()
```
