```python
from uproot_io import Events, View
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
events = Events(r"C:\Users\andre\Desktop\Physics Project\bigdata\CheatedRecoFile_1.root")
```

```python
allevents = np.arange(max(events.event_number))
```

```python
truth = []
mains = []
parents = []
for n in allevents:
    idx = np.where(events.event_number == n)[0]
    main = events.mc_pdg
    print(main)
    for i in idx:
        main = events.mc_pdg[i]
        parent = events.reco_parent_index[i]
        if main < 250 and main > -250 and main != 0:
            mains.append(main)
        if parent < 50:
            parents.append(parent)
        
        x = events.reco_hits_x_w[i]
        z = events.reco_hits_w[i]
        pdg = events.mc_pdg[i]
        purity = events.purity[i]
        complete = events.completeness[i]
        if len(x) > 10 and pdg != 0 and purity > 0.75 and complete > 0.75 and main != 0:
            if main == 13. or main == -13.:
                truth.append(0)
            if main == 11. or main == -11.:
                if parent == 0:
                    truth.append(1)
                else:
                    truth.append(2)
            else:
                truth.append(2)
```
